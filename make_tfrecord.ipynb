```python
from preprocess import *
import librosa
import numpy as np
import tensorflow as tf
import os
tf.enable_eager_execution()
```

```python
def _bytes_feature(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))
def _floats_feature(value):
    return tf.train.Feature(float_list=tf.train.FloatList(value=value))
    #return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))
def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
```

```python
max_len =11
def parse_func(s_example):
    features = {
                #'wave': tf.VarLenFeature(tf.float32),
                #'wave': tf.FixedLenFeature((22050,), tf.float32),
                #'sr': tf.FixedLenFeature((), tf.int64),
                #'mfcc':tf.VarLenFeature(tf.float32),
                'mfcc2':tf.FixedLenFeature((20,11,1),tf.float32),
                'n_mfcc':tf.FixedLenFeature((), tf.int64),
                'word': tf.FixedLenFeature((), tf.string)
                }
    example = tf.parse_single_example(s_example, features=features)
    return example

def imgSerialization (file_path,label):
    mfcc2= wav2mfcc(file_path)
    n_mfcc=mfcc2.shape[0]
    example = tf.train.Example(features=tf.train.Features(feature={
        'word': _bytes_feature(label.encode('utf-8')),
        #'sr': _int64_feature(sr),
        #'mfcc':_floats_feature(mfcc.reshape(-1)),
        'mfcc2':_floats_feature(mfcc2.reshape(-1)),
        'n_mfcc':_int64_feature(n_mfcc),
        #'wave': _floats_feature(wave)
        }))
    return example.SerializeToString()
```

```python
DATA_PATH = "./data/"
for label in os.listdir(DATA_PATH):
    directory = os.path.join(DATA_PATH,label)
    writer = tf.python_io.TFRecordWriter(label+'.tfrecords')
    files = tf.data.Dataset.list_files(file_pattern=directory +'/*.wav')
    for file in files:
        s_example = imgSerialization(file.numpy().decode('utf-8'),label)
        writer.write(s_example)
    writer.close()
```

```python
data = tf.data.TFRecordDataset(['bed.tfrecords','cat.tfrecords','happy.tfrecords']).shuffle(20000).map(parse_func).take(100)
#iterator = data.make_one_shot_iterator()
```
